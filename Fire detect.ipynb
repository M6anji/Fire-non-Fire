```python
import cv2

face_decetor = cv2.CascadeClassifier( cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
video_capture = cv2.VideoCapture(0)
def Faces(vid):
    gray_image = cv2.cvtColor(vid, cv2.COLOR_RGB2GRAY)
    faces = face_decetor.detectMultiScale(gray_image)
    for (x, y, w, h) in faces:
        cv2.rectangle(vid, (x, y), (x + w, y + h), (0, 255, 255),3)
    return faces
while True:

    result, video_frame = video_capture.read()  # read frames from the video
    if result is False:
        break  # terminate the loop if the frame is not read successfully

    faces = Faces(video_frame)  # apply the function we created to the video frame

    cv2.imshow(
        "My Face Detection Project", video_frame
    )  # display the processed frame in a window named "My Face Detection Project"

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

video_capture.release()
cv2.destroyAllWindows()
```

```python
model = models.load_model("mlo.h5")
vid = cv2.VideoCapture(0)
while (True):
    result , frame = vid.read()
    
    frame_resized = cv2.resize(frame, (32, 32))
    frame_flatten = frame_resized.flatten()
    frame_flatten = np.expand_dims(frame_flatten, axis=0)
    frame_flatten = frame_flatten / 255.0
    
    output = model.predict(frame_flatten)
    max_output = np.argmax(output)
    if max_output == 0:
        label = 'Fire'
        color = (0, 255, 0)  
    else:
        label = 'Non_Fire'
        color = (0, 0, 255)
    
    cv2.putText(frame, label, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 2,color , 2 , cv2.LINE_AA)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
    cv2.imshow('Fire_Non-fire', frame)
vid.release()
cv2.destroyAllWindows()
```
